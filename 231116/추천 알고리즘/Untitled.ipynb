```python
def calc_age(birthdate):
    today = datetime.datetime.today()
    birthdate = datetime.datetime.strptime(birthdate, '%Y-%m-%d')
    age = today.year-birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age
```

```python
df['age'] = df['birth_date'].apply(calc_age)
user_age = df[df['username'] == 'rcFp2vNap']['age'].values[0]
user_salary = df[df['username'] == 'rcFp2vNap']['salary'].values[0]
user_age_balance = df[df['username'] == 'rcFp2vNap']['balance'].values[0]
df['age_diff'] = abs(df['age'] - user_age)
selected_columns = ['username', 'salary', 'balance', 'age_diff']
result_df = df.loc[:, selected_columns]
result_df.sort_values(by=['age_diff', 'salary', 'balance'])
```
