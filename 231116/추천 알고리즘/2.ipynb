```python
import pandas as pd
import sqlite3
import datetime
```

```python
def calc_age(birthdate):
    today = datetime.datetime.today()
    birthdate = datetime.datetime.strptime(birthdate, '%Y-%m-%d')
    age = today.year-birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age
```
